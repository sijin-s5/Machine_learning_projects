```python
x=df.iloc[:,:-1].values #x always 2D
x.ndim
x
```

```python
y=df.iloc[:,-1].values   #y always 1D
y.ndim
y
```

```python
#Convert data into training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)#Random state=we set a values to perform model evaluation in every system equally
                                                                                  #commenly used 1,0,42
x_train
```

```python
#Normalisation or scaling
     #All inputs in a dataset are transformed into perticular range.
#2 Methods

#1.Standard scalar
#2. Min-Max scalar

#1.Standard scalar
  # Works in mathematical equation
#z=(x-u)/s
#z=output
#X=input 
#u=Mean of training data
#s=Standard deviation of training data
#eg:
# Pregnency 
#z=(6-mean of training data )/stan...... 6=First data in the pregnency column then next ....
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)  #fit=collecting x_train because we collect only x_train datas
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train
```

```python
x_test
```

```python
#Model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Passing random values
print(knn.predict(scalar.transform([[3,110,76,32,5,40,0.32,45]])))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score,classification_report
cr=classification_report(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
#Hyper parameter tuning
from sklearn.model_selection import GridSearchCV   #CV=cross validation = use 10 times folding for good accuracy
model=KNeighborsClassifier()
parameter={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(model,parameter,cv=10,scoring='accuracy')
clf.fit(x_train,y_train)  #k=3 ,w:uni

```

```python
print(clf.best_params_)
```
