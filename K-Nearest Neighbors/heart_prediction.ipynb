```python
df=df.dropna()
df['restecg'].unique()
```

```python
df['trestbps'].unique()
```

```python
# x=df['trestbps'].mean()
# df['trestbps'].fillna(x,inplace=True)

#Easy method
a=df['ca'].mode
df['ca'].fillna(a,inplace=True)

df['trestbps']=df['trestbps'].fillna(df['trestbps'].mean())
df['restecg']=df['restecg'].fillna(df['restecg'].mode())
df['thalach']=df['thalach'].fillna(df['thalach'].mean())
df['thal']=df['thal'].fillna(df['thal'].mode())
df.isna().sum()
```

```python
df.drop(columns='ca',inplace=True)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)  #fit=collecting x_train because we collect only x_train datas
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
print(type(x_train))
print(x_train[:5])

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
